```python
# Создаем класс 
class MyClass:
    # Метод для присваивания значения полю экземпляра класса
    def set(self, n):
        """ Метод для присваивания значения полю 
            экземпляра класса self.number=n
        """
        print("Внимание! Присваивается значение!")
        # Полю присваивается значение
        self.number=n
    # Метод для считывания значения поля экземпляра класса
    def get(self):
        """ Метод для считывания значения экземпляра класса"""
        # Отображаем значение поля
        print("Значение поля:", self.number)
# Создаем экземпляр класса
obj=MyClass()
# Вызывается метод экземпляра класса и полю экземпляра 
# класса присваивается значение
obj.set(1001)
# Вызывается метод экземпляра класса и отображается
# значение поля экземпляра
obj.get()
```

## Конструктор и деструктор экземпляра класса

                                                        Как говорит наш дорогой шеф, в нашем деле
                                                        главное — этот самый реализм ,
                                                        из к/ф "Бриллиантоваярука"
При создании класса можно описать специальный метод, который называется конструктором экземпляра класса (иногда также называют методом
инициализации). Этот метод автоматически вызывается при создании экземпляра класса. Рецепт создания конструктора очень простой: необходимо создать метод с названием__ i n i t __ () (два символа подчеркивания в
начале и два символа подчеркивания в конце). Аргументов у конструктора может быть сколько угодно - но не меньше одного (ссылка на экземпляр,
при создании которого вызывается конструктор).

```python
# Создаем класс
class MyClass:
    # Конструктор
    def __init__ (self):
        # Присваивается значение полю
        self.number=0
        # Отображается сообщение
        print("Создан экземпляр класса!")
# Создается экземпляр класса
obj=MyClass()
# Проверяем значение поля экземпляра класса
print("Значение поля:", obj.number)
second=MyClass()
```

У конструктора может быть несколько аргументов (во всяком случае, больше одного). В этом случае при создании экземпляра класса конструктору
нужно передать необходимые аргументы. 

```python
# Сооздаем класс
class MyClass:
    
    # Метод для присваивания значения полю
    def set(self, n):
        # Полю number присваивается значение
        self.num=n
    # Метод для отображения значения поля
    def get(self):
        # Отображает значение поля number
        print("Значение поля:", self.num)
    # Конструктор с двумя аргументами.
    # У второго аргумента есть значение по умолчанию
    def __init__(self, n=0):
        # Вызывает метод set() для присваивания значения полю number
        self.set(n)
        # Отображает сообщение
        print("Создан экземпляр класса.")
        # Вызывается метод get() для отображения значения поля number
        self.get()
a=MyClass()
b=MyClass(100)

```

Метод с названием `__d e l__()` (два символа подчеркивания в начале и
два - в конце) автоматически вызывается при удалении экземпляра класса из памяти. Этот метод принято называть деструктором. У деструктора
один и только один аргумент (ссылка на экземпляр класса `s e l f` ) .

```python
# Класс с конструктором и деструктором
class MyClass:
    # Конструктор
    def __init__(self):
       print("Всем привет!")
    # Деструктор
    def __del__(self):
       print("Всем пока!")
print("Проверяем работу деструктора.")
# Создаем экземпляр класса
obj=MyClass()
print("Экземпляр класса создан. Удаляем его.")
# Удаляем экземпляр класса
del obj
print("Выполнение программы завершено")
```
